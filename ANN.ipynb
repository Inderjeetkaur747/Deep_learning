```python
ds = pd.read_csv('Churn_Modelling.csv')
ds.head(5)
```

```python
ds.drop(columns = ["RowNumber","CustomerId", "Surname", "Geography", "Gender"], inplace = True)
```

```python
ds.head(4)
```

```python
input_data = ds.iloc[:,:-1]
output_data = ds.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data), columns = input_data.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size = 0.2, random_state = 42)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from keras.callbacks import EarlyStopping
from keras.regularizers import L2
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6, input_dim = 8, activation = 'relu', kernel_regularizer= L2(0.01)))
# ANN.add(Dropout(0.5))
# ANN.add(BatchNormalization())
ann.add(Dense(4,activation = 'relu'))
# ANN.add(Dropout(0.5))
# ANN.add(BatchNormalization())
ann.add(Dense(2, activation ='relu'))
ann.add(Dense(1, activation = 'sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss= 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = ann.fit(x_train, y_train, epochs = 40, batch_size= 100, validation_split = 0.2, callbacks= EarlyStopping())
```

```python
prd = ann.predict(x_test)
```

```python
prd_data = []
for i in prd:
  if i > 0.5:
    prd_data.append(1)
  else:
    prd_data.append(0)
```

```python
prd1 = ann.predict(x_train)
```

```python
prd1_data = []
for i in prd1:
  if i > 0.5:
    prd1_data.append(1)
  else:
    prd1_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prd_data)*100
```

```python
accuracy_score(y_train, prd1_data)*100
```
